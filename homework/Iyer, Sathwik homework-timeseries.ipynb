3. (3 pts) Bandpass filter the recording from 5-40 Hz. Plot the filter kernel. Also plot the original recording overlaid with the filtered version. Also plot the power spectral density of the original recording overlaid with the power spectral density for the filtered version.

```python
cutoffFreq_Hz = [5, 40]
bandpass = signal.firwin(201, cutoffFreq_Hz, pass_zero=False, nyq=nyquistFreq_Hz)
filtered_data = np.convolve(data, bandpass, 'same')

plt.subplot(2,1,1)
plt.plot(bandpass, '.-')
plt.title('bandpass filter')
```

```python
plt.subplot(2,1,2)
plt.plot(filtered_data, '.-')
plt.plot(data, '.-')
plt.xlabel('Time (ms)')
plt.title('Filtered and Original Data Overlaid');
```

4. (3 pts) Plot the fast Fourier transform of the original and filtered data (overlay both on the same plot). Zoom in on the y-axis so that your filtered FFT takes up approximately half the height of the axes.

```python
sampleInterval_ms = time_ms[1] - time_ms[0]
fft_data = np.fft.fft(data)
fft_bp_data = np.fft.fft(filtered_data)

freq_kHz_data = np.fft.fftfreq(len(fft_data), sampleInterval_ms)
freq_kHz_bp = np.fft.fftfreq(len(fft_bp_data), sampleInterval_ms)

ind = np.where(freq_kHz_data >= 0)[0]

plt.axis([0,.1,0,200])
plt.plot(freq_kHz_data[ind], np.abs(fft_data[ind]), label = "original")
plt.plot(freq_kHz_bp[ind], np.abs(fft_bp_data[ind]), label = "filtered")
plt.xlabel('Frequency (kHz)')
plt.title('FFT - Amount of each frequency component in data');
plt.legend();
```
