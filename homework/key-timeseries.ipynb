1. (3 pts) What are the sample and Nyquist frequencies for this recording in Hz? Pay attention to your units!

```python
sampleInterval_s = (time_ms[1] - time_ms[0]) / 1000
sampleFreq_Hz = 1 / sampleInterval_s
nyquistFreq_Hz = sampleFreq_Hz / 2

print(f"Sample frequency = {sampleFreq_Hz :.2f} Hz")
print(f"Nyquist frequency = {nyquistFreq_Hz :.2f} Hz")
```

2. (3 pts) Properly downsample to every 10th data point. Plot the original recording overlaid with the downsampled version.

```python
from scipy import signal
import matplotlib.pyplot as plt

subtime_ms = time_ms[::10]
subdata = signal.decimate(data, 10)

plt.figure(figsize=(15,5))
plt.plot(time_ms, data, label='original')
plt.plot(subtime_ms, subdata, lw=2, label='downsampled')
plt.xlabel('Time (ms)')
plt.legend();
```

3. (3 pts) Bandpass filter the recording from 5-40 Hz. Plot the filter kernel. Also plot the original recording overlaid with the filtered version. Also plot the power spectral density of the original recording overlaid with the power spectral density for the filtered version.

```python
bandpass = signal.firwin(201, [5, 40], pass_zero=False, nyq=nyquistFreq_Hz)

plt.plot(bandpass, '.-')
plt.xlabel('Time Pts')
plt.title('Bandpass Filter');
```

```python
filteredData = np.convolve(data, bandpass, 'same')

plt.figure(figsize=(15,5))
plt.plot(time_ms, data, label='original')
plt.plot(time_ms, filteredData, label='filtered', linewidth=3)
plt.xlabel('Time (ms)')
plt.legend();
```

4. (3 pts) Plot the fast Fourier transform of the original and filtered data (overlay both on the same plot). Zoom in on the y-axis so that your filtered FFT takes up approximately half the height of the axes.

```python
fft = np.fft.fft(data)
freq = np.fft.fftfreq(len(fft), sampleInterval_s)
ind = np.where(freq >= 0)[0]
plt.plot(freq[ind], np.abs(fft[ind]), label='original')
plt.xlabel('Frequency (Hz)')

fft = np.fft.fft(filteredData)
freq = np.fft.fftfreq(len(fft), sampleInterval_s)
ind = np.where(freq >= 0)[0]
plt.plot(freq[ind], np.abs(fft[ind]), label='filtered')
plt.ylim([0, 200])
plt.legend();
```

**Extra Credit** (3 pts) Plot the power spectral density for the original and filtered data  (overlay both on the same plot).

```python
plt.psd(data, Fs=sampleFreq_Hz, label='original');
plt.psd(filteredData, Fs=sampleFreq_Hz, label='filtered');
plt.xlabel('Frequency (Hz)')
plt.legend();
```
